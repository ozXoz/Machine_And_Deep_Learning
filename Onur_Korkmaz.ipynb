```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread('/content/data/onur3.jpg')
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
result = DeepFace.analyze(img1,enforce_detection=False, actions=['emotion'])
```

```python
print("Predicted Age is : " , result['emotion'])
```
